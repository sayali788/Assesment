```python


def generate_historical_quiz_recommendations(historical_quiz):

    weak_accuracy_scores = historical_quiz[historical_quiz['accuracy'] < 0.60].copy()  
    
    
    weak_accuracy_scores.loc[:, 'quiz_topic'] = weak_accuracy_scores['quiz'].apply(lambda x: x.get('topic', 'Unknown'))
   
    weak_accuracy_types = weak_accuracy_scores['type'].unique()
    
    
    recommendations = {
        "weak_accuracy_topics": weak_accuracy_scores['quiz_topic'].tolist(),
        "weak_accuracy_types": weak_accuracy_types.tolist(),
        "general_advice": "Focus on improving accuracy in these topics and question types. Consider practicing similar questions to boost understanding."
    }
    
    return recommendations

historical_quiz_recommendations = generate_historical_quiz_recommendations(historical_quiz)
print("Historical Quiz Recommendations:")
print(historical_quiz_recommendations)
```

```python
#Analyze and define the student persona based on patterns in the data.


def analyze_student_persona(historical_quiz):
    
    historical_quiz['duration_cleaned'] = pd.to_datetime(historical_quiz['duration'], format='%H:%M', errors='coerce').dt.strftime('%H:%M:%S')
    historical_quiz['duration_minutes'] = pd.to_timedelta(historical_quiz['duration_cleaned']).dt.total_seconds() / 60

    high_performers = historical_quiz[historical_quiz['accuracy'] > 0.85]

    low_performers = historical_quiz[historical_quiz['accuracy'] < 0.60]
    

    high_performer_topics = high_performers['quiz'].apply(lambda x: x.get('topic', 'Unknown')).value_counts()
    low_performer_topics = low_performers['quiz'].apply(lambda x: x.get('topic', 'Unknown')).value_counts()
    
   
    high_performer_persona = {
        "persona_type": "High Performer",
        "accuracy": high_performers['accuracy'].mean(),
        "topics_strong_in": high_performer_topics.head(3).index.tolist(),
        "average_score": high_performers['score'].mean(),
        "average_duration": high_performers['duration_minutes'].mean(),
        "speed": high_performers['speed'].mean(),
        "recommendation": "Maintain high performance in your strong topics, challenge yourself with more complex topics."
    }
    
    low_performer_persona = {
        "persona_type": "Low Performer",
        "accuracy": low_performers['accuracy'].mean(),
        "topics_struggling_with": low_performer_topics.head(3).index.tolist(),
        "average_score": low_performers['score'].mean(),
        "average_duration": low_performers['duration_minutes'].mean(),
        "speed": low_performers['speed'].mean(),
        "recommendation": "Focus on improving accuracy in your weak topics and practice more frequently."
    }
    
    return high_performer_persona, low_performer_persona

high_performer_persona, low_performer_persona = analyze_student_persona(historical_quiz)


print("High Performer Persona:", high_performer_persona)
print("Low Performer Persona:", low_performer_persona)
```

```python
#Here new name given for student persona analyzed student data to get high and low persona of student
def analyze_student_persona(historical_quiz):
  
    historical_quiz['duration_cleaned'] = pd.to_datetime(historical_quiz['duration'], format='%H:%M', errors='coerce').dt.strftime('%H:%M:%S')
    historical_quiz['duration_minutes'] = pd.to_timedelta(historical_quiz['duration_cleaned']).dt.total_seconds() / 60


    high_performers = historical_quiz[historical_quiz['accuracy'] > 0.85]
    
    low_performers = historical_quiz[historical_quiz['accuracy'] < 0.60]
    
    
    high_performer_topics = high_performers['quiz'].apply(lambda x: x.get('topic', 'Unknown')).value_counts()
    low_performer_topics = low_performers['quiz'].apply(lambda x: x.get('topic', 'Unknown')).value_counts()
  
    high_performer_persona = {
        "persona_type": "Quiz Master",
        "accuracy": high_performers['accuracy'].mean(),
        "topics_strong_in": high_performer_topics.head(3).index.tolist(),
        "average_score": high_performers['score'].mean(),
        "average_duration": high_performers['duration_minutes'].mean(),
        "speed": high_performers['speed'].mean(),
        "strengths": "Excellent in complex topics, fast responders.",
        "recommendation": "Continue excelling in your strong topics and dive into more challenging quizzes to push your limits. You are a master in the making!"
    }
    
    low_performer_persona = {
        "persona_type": "Struggling Star",
        "accuracy": low_performers['accuracy'].mean(),
        "topics_struggling_with": low_performer_topics.head(3).index.tolist(),
        "average_score": low_performers['score'].mean(),
        "average_duration": low_performers['duration_minutes'].mean(),
        "speed": low_performers['speed'].mean(),
        "weaknesses": "Struggling with certain topics, possibly due to low speed or accuracy.",
        "recommendation": "Focus on improving your weak areas and speed through more focused practice. You have the potential to shine with some extra effort!"
    }
    
    return high_performer_persona, low_performer_persona


high_performer_persona, low_performer_persona = analyze_student_persona(historical_quiz)


print("High Performer Persona:", high_performer_persona)
print("Low Performer Persona:", low_performer_persona)
```
